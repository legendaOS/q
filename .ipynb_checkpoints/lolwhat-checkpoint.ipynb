```python
import pandas
```

```python
d = pandas.read_excel('2018.xlsx', engine='openpyxl')
```

```python
temp = []
buf = []
c = 0
for i in d['T']:
  buf.append(i)
  c += 1
  if c == 8:
    temp.append(buf)
    buf = []
    c = 0
```

```python
train = temp[0:17]
```

```python
temp
```

```python
train_set = []

for i in range(len(train)- 6):
  buf = []
  buf.append(train[i : i+5])
  train_set.append(buf[0])
```

```python
y_train_set = [i for i in train[5:-1]]
```

```python
train_set
```

```python
y_train_set
```

```python
x_train_set  = []
for i in train_set:
  buf = []
  for j in i:
    for k in j:
      buf.append(k)
  x_train_set.append(buf)
```

```python
print(x_train_set[0])
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, SimpleRNN, LSTM, Flatten
```

```python
model = keras.Sequential()
model.add(Dense(input_shape = (40,), units=20, activation='linear'))
model.add(Dense(units=20, activation='linear'))
model.add(Dense(units=8, activation='linear'))

model.compile(optimizer=tf.optimizers.Adam(0.1), loss='mse')
```

```python
model.fit(x_train_set, y_train_set, epochs = 700, verbose=0)
```

```python
model.predict([x_train_set[0]])
```

```python
s_sr = {0:[], 1:[], 2:[], 3:[]}
for pred in sr:
  sred = 0
  count = 0
  for arr in sr[pred]:
    for elm in arr:
      sred += elm
      count += 1
  sred /= count
  s_sr[pred].append(sred)
  s_sr[pred].append(count // 8)
```
