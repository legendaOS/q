чтение данных о погоде за 2019 год

```python
import pandas as pd
```

```python
d = pd.read_excel('2019.xlsx', engine='openpyxl')
```

Созднаие обучающей выборки

```python
temper = []
count = 0
buf = []
for el in d['T']:
    buf.append(float(el))
    count += 1
    if count == 8:
        temper.append(buf)
        buf = []
        count = 0
```

Обучение

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters =4)
```

```python
km.fit(temper)
```

тестовая выборка

```python
from random import randint as randint
```

```python
t_test = []
for i in range(10):
    buf_c = randint(0, len(temper) - 1)
    t_test.append(temper[buf_c])
```

```python
t_test
```

```python
km.predict(t_test)
```

Проверка на данных 2020 года

```python
d_2020 = pd.read_excel('2018.xlsx', engine='openpyxl')
```

```python
temper2020 = []
count = 0
buf = []
for el in d_2020['T']:
    buf.append(float(el))
    count += 1
    if count == 8:
        temper2020.append(buf)
        buf = []
        count = 0
```

```python
prediction = km.predict(temper2020)
```

```python
elem = {}
for i in prediction:
    if not i in elem:
        elem[i] = 1
    else:
        elem[i] += 1
```

```python
def cl_arr(a):
    buf = [0,0,0,0]
    for i in a:
        buf[i] += 1
    return buf
```
